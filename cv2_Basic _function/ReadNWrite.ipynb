## Importing libraries

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

## Set webcam for capturing frames

```python
webcam = cv2.VideoCapture(0)
ret, frame = webcam.read()

# release function for release the webcam 
webcam.release()
```

## Save captured frame as image 

```python
print(ret)
if ret:
    # imwrite function save the frame/Image
    cv2.imwrite('face.png', frame)
```

## Show captured frame

```python
# open a new thread to manage the external cv2 interaction
cv2.startWindowThread()
cv2.namedWindow("Face", cv2.WINDOW_NORMAL)

# show the image in window
cv2.imshow('Face', frame)
cv2.waitKey()
cv2.destroyAllWindows()
```

## plot image using plt

```python
# plot image using plt
plt.imshow(frame)
plt.show()
# It is ugly bcz it is in BGR
```

## convert image from BGR to RGB

```python
# convert image from BGR to RGB
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(frame)
plt.show()
```
